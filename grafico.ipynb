```python
import matplotlib.pyplot as plt

# Dados para o gráfico com base nas informações do relatório da ESET
categories = [
    "Empresas alvo de ransomware",
    "Organizações com incidentes de segurança",
    "Ataques com exploits em vulnerabilidades antigas",
    "Orçamento de cibersegurança insuficiente"
]
values = [23, 30, 81, 62]  # valores percentuais

# Criar o gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(categories, values, color=['#ff6f61', '#6baed6', '#74c476', '#fd8d3c'])
ax.set_xlabel("Percentual (%)", fontsize=12)
ax.set_title("Estatísticas de Segurança Cibernética na América Latina (ESET 2024)", fontsize=14)
ax.grid(True, linestyle='--', alpha=0.7)

# Exibir o gráfico
plt.show()
```
