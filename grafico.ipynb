```python
import matplotlib.pyplot as plt

# Dados para o grfico com base nas informaes do relatrio da ESET
categories = [
    "Empresas alvo de ransomware",
    "Organizaes com incidentes de segurana",
    "Ataques com exploits em vulnerabilidades antigas",
    "Oramento de cibersegurana insuficiente"
]
values = [23, 30, 81, 62]  # valores percentuais

# Criar o grfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(categories, values, color=['#ff6f61', '#6baed6', '#74c476', '#fd8d3c'])
ax.set_xlabel("Percentual (%)", fontsize=12)
ax.set_title("Estatsticas de Segurana Ciberntica na Amrica Latina (ESET 2024)", fontsize=14)
ax.grid(True, linestyle='--', alpha=0.7)

# Exibir o grfico
plt.show()
```
