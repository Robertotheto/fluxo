```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the graph (years and corresponding threat data growth)
years = np.arange(2015, 2024)
threats_growth = [100, 130, 170, 220, 280, 360, 460, 600, 750]  # growth of cyber threats (in percentage)
log_data_growth = [150, 180, 240, 310, 400, 500, 650, 820, 1000]  # growth of log data generated (in TB)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot both lines on the graph
ax.plot(years, threats_growth, label='Crescimento de Ameaas Cibernticas (%)', color='red', marker='o', linestyle='--')
ax.plot(years, log_data_growth, label='Crescimento de Dados de Logs Gerados (TB)', color='blue', marker='s', linestyle='-')

# Add title and labels
ax.set_title('Crescimento de Ameaas Cibernticas e Dados de Logs (2015-2023)', fontsize=14)
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Crescimento (%) ou Dados (TB)', fontsize=12)

# Add grid and legend
ax.grid(True)
ax.legend()

# Show the plot
plt.show()
```
