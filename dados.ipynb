```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Criando uma figura
# Adjusting the arrows to ensure correct direction from sources to SIEM, SIEM to IA correlation, and IA correlation to final output.

# Create figure
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Drawing source boxes
sources = [
    ("Logs de Rede", 1, 8),
    ("Logs de Sistema", 1, 6.5),
    ("Autenticao", 1, 5),
    ("Eventos Endpoint", 1, 3.5)
]

# Main process boxes
main_boxes = [
    ("Processamento pelo SIEM", 4, 6.5),
    ("Correlao de Eventos com IA", 7, 5)
]

# Final output box
output_box = ("Alertas e Relatrios de Segurana", 7, 2.5)

# Drawing sources
for label, x, y in sources:
    ax.add_patch(FancyBboxPatch((x, y), 2, 0.7, boxstyle="round,pad=0.3", fc="lightblue", ec="black"))
    ax.text(x + 1, y + 0.35, label, ha="center", va="center", fontsize=10)

# Drawing main processing boxes
for label, x, y in main_boxes:
    ax.add_patch(FancyBboxPatch((x, y), 2.5, 1, boxstyle="round,pad=0.3", fc="lightgreen", ec="black"))
    ax.text(x + 1.25, y + 0.5, label, ha="center", va="center", fontsize=10)

# Final output box
ax.add_patch(FancyBboxPatch((output_box[1], output_box[2]), 2.5, 1, boxstyle="round,pad=0.3", fc="salmon", ec="black"))
ax.text(output_box[1] + 1.25, output_box[2] + 0.5, output_box[0], ha="center", va="center", fontsize=10)

# Corrected arrow directions
arrow_props = dict(facecolor='black', arrowstyle="->")
for _, x, y in sources:
    ax.annotate('', xy=(4, 7.5), xytext=(x + 2, y + 0.35), arrowprops=arrow_props)  # sources to SIEM
ax.annotate('', xy=(7, 6), xytext=(6.5, 6), arrowprops=arrow_props)  # SIEM to IA correlation
ax.annotate('', xy=(7, 3.5), xytext=(7, 4.5), arrowprops=arrow_props)  # IA to final alert

# Titles and final setup
ax.set_title("Fluxo de Dados em um Sistema SIEM com IA para Correlao de Eventos", fontsize=14)
ax.axis("off")

plt.show()
```
