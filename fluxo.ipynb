```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define the figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Create flowchart elements with text
ax.text(0.5, 0.9, "Coleta de Logs", fontsize=12, ha='center', bbox=dict(facecolor='lightblue', boxstyle="round,pad=0.3"))
ax.text(0.5, 0.75, "Processamento pelo SIEM", fontsize=12, ha='center', bbox=dict(facecolor='lightgreen', boxstyle="round,pad=0.3"))
ax.text(0.5, 0.6, "Correlação de Eventos", fontsize=12, ha='center', bbox=dict(facecolor='lightyellow', boxstyle="round,pad=0.3"))
ax.text(0.5, 0.45, "Análise Preditiva por IA", fontsize=12, ha='center', bbox=dict(facecolor='orange', boxstyle="round,pad=0.3"))
ax.text(0.5, 0.3, "Identificação de Ameaças", fontsize=12, ha='center', bbox=dict(facecolor='salmon', boxstyle="round,pad=0.3"))
ax.text(0.5, 0.15, "Envio Automatizado de Alertas (via mensagem)", fontsize=12, ha='center', bbox=dict(facecolor='lightcoral', boxstyle="round,pad=0.3"))

# Draw arrows between boxes
arrow_props = dict(facecolor='black', arrowstyle="->")
ax.annotate('', xy=(0.5, 0.82), xytext=(0.5, 0.88), arrowprops=arrow_props)
ax.annotate('', xy=(0.5, 0.67), xytext=(0.5, 0.73), arrowprops=arrow_props)
ax.annotate('', xy=(0.5, 0.52), xytext=(0.5, 0.58), arrowprops=arrow_props)
ax.annotate('', xy=(0.5, 0.37), xytext=(0.5, 0.43), arrowprops=arrow_props)
ax.annotate('', xy=(0.5, 0.22), xytext=(0.5, 0.28), arrowprops=arrow_props)

# Set axis limits and remove the axis
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Show the diagram
plt.show()
```
